```python
# imports and load data
import pandas as pd
%matplotlib inline

df = pd.read_csv('store_data.csv', index_col='week')
df
```

```python
# explore data
sum(df.duplicated())
```

```python
df.info()
```

```python
# total sales for the last month
last_month=df.iloc[-4:, :]
last_month

last_month['total sales'] = last_month['storeA']+last_month['storeB']+last_month['storeC']+last_month['storeD']+last_month['storeE']

#last_month

print(last_month['storeA'].sum(), last_month['storeB'].sum(), last_month['storeC'].sum(), last_month['storeD'].sum(), last_month['storeE'].sum())
```

```python
# average sales
avgA = df['storeA'].mean()
avgB = df['storeB'].mean()
avgC = df['storeC'].mean()
avgD = df['storeD'].mean()
avgE = df['storeE'].mean()
print(avgA, avgB, avgC, avgD, avgE)
```

```python
# sales on march 13, 2016

sale_df = df.loc['2016-03-13', :]
sale_df
```

```python
# worst week for store C
lowestC = df.loc[:, 'storeC'].min()
y = df.loc[df['storeC'] == lowestC]
y
```
